```python
import pandas as pd
rf_csv = r'C:\Users\Acer\Desktop\3dVS\RFcsv\volume_data5.csv'
volume_table = pd.read_csv(rf_csv, index_col=0)
volume_table
```

```python
volume_table = volume_table.dropna(subset=['volume_diff'])
volume_table
```

```python
volume_table = volume_table.drop(columns='volume_diff')
volume_table
```

```python
y = volume_table['change']
x=volume_table.drop(['change'],axis=1)
```

```python
y.value_counts()
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# Recursive Feature Elimination
model = RandomForestClassifier()
rfe = RFE(estimator=model, n_features_to_select=10)
X_rfe = rfe.fit_transform(x, y)
```

```python
selected_feature_indices_rfe = rfe.get_support(indices=True)
selected_feature_names_rfe = x.columns[selected_feature_indices_rfe].tolist()
selected_feature_names_rfe
```

```python
x_wrapper=x[selected_feature_names_rfe]
```

# Training

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train, X_test , y_train, y_test = train_test_split(x_wrapper, y, test_size=0.2, random_state=42)
```

```python
y_train.value_counts()
```

```python
# Random Forest

rf_classifier = RandomForestClassifier(n_estimators=200, random_state=1)
rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test, y_pred))
```
